```python
#importamos las librerias a usar

import requests
import pandas as pd
from datetime import datetime
```

```python
#Extraemos la data desde el API haciendo uso de requests en formato json

API = "https://api.covidtracking.com/v1/us/daily.json"
response = requests.get(API).json()
```

```python
# le damos el formato de dataframe a nuestro json con ayuda de pandas y guardamos esta data ordenada por la fecha

data = pd.json_normalize(response)

df = data.sort_values(by="date").reset_index(drop=True)
```

```python
#modificamos las columnas que hacen referencia a una fecha y le damos el formato correspondiente

df['lastModified'] = df['lastModified'].str.replace('T24:00:00Z', 'T00:00:00Z')
df['lastModified'] = pd.to_datetime(df['lastModified'], format='%Y-%m-%dT%H:%M:%SZ').dt.date


df['dateChecked'] = df['dateChecked'].str.replace('T24:00:00Z', 'T00:00:00Z')
df['dateChecked'] = pd.to_datetime(df['dateChecked'], format='%Y-%m-%dT%H:%M:%SZ').dt.date


df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
#filtramos la data

df = df[df['positive'] > 0]

#Seleccionamos las columnas a usar

data_columns = ['date', 'positive', 'negative', 'pending',
       'hospitalizedCurrently', 'hospitalizedCumulative', 'inIcuCurrently',
       'inIcuCumulative', 'onVentilatorCurrently', 'onVentilatorCumulative',
       'dateChecked', 'death', 'hospitalized', 'totalTestResults',
       'lastModified', 'total', 'posNeg', 'deathIncrease',
       'hospitalizedIncrease', 'negativeIncrease', 'positiveIncrease',
       'totalTestResultsIncrease']

#generamos un id para nuestro dataframe
df = df[data_columns].reset_index(drop=True).reset_index()
df.rename(columns={'index': 'id'}, inplace=True)
df['id'] = df['id'] + 1
```
